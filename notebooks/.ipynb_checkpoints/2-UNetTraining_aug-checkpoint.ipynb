```python
def removeFileinType(lakeType,patchType,fileName,annName):
    output_dir=os.path.join(config.base_dir,'patchsReshape\\'+patchType+'\\type{}'.format(lakeType))
    new=os.path.join(output_dir,fileName)
    old=os.path.join(patch_output,fileName)
    shutil.copy(old, new)
    new1=os.path.join(output_dir,annName)
    old1=os.path.join(patch_output,annName)
    shutil.copy(old1, new1)
    
def removeFile(lakeType,patchType,fileName,annName):
    output_dir=os.path.join(config.base_dir,'patchsReshape\\'+patchType)
    new=os.path.join(output_dir,fileName)
    old=os.path.join(patch_output,fileName)
    shutil.copy(old, new)
    new1=os.path.join(output_dir,annName)
    old1=os.path.join(patch_output,annName)
    shutil.copy(old1, new1)
```

# 老方法

```python
def imageAugmentationWithIAA():
    sometimes = lambda aug, prob=0.5  : iaa.Sometimes(prob, aug)
    seq = iaa.Sequential([
        # Basic aug without changing any values
        iaa.Fliplr(0.5),  # horizontally flip 50% of all images 左右翻转
        iaa.Flipud(0.5),  # vertically flip 50% of all images 上下翻转
        sometimes(iaa.Crop(percent=(0, 0.1))),  # random crops
    ],
        random_order=True)
    return seq
        
# Return all training and label images, generated sequentially with the given step size
def sequential_patches(frames,input_image_channel,annotation_channel, patch_size, step_size,augmenter):
      
        seq = imageAugmentationWithIAA()
        patches = []
#         patches_lonlat_ul = [] # new add transform
        for frame in frames:
            ps=frame.sequential_patches(patch_size, step_size,0)
            patches.extend(ps)
            
        data = np.array(patches)
        img = data[..., input_image_channel]
        y = data[..., annotation_channel] 
        ann = y[...,[0]]
        ann_joint=ann
        
        #imageAugmentation
        if augmenter == 'iaa':   #augmenter = 'iaa'  
                seq_det = seq.to_deterministic()
                img = seq_det.augment_images(img)
    
                # y would have one channel, i.e. annotations. We need to augmenty for operations such as crop and transform
                ann_joint = seq_det.augment_images(ann_joint) 
                ann_joint=ann_joint[...,[0]]
        return (img, ann_joint) 
```

```python
#to balance the patches numbers of each type, use different step size and use data augmentation strategy.
train_img_list=[]
train_ann_list=[]

for i in range(1,5):
    frames=[]
    print("========type"+str(i)+"========")
    path_to_write=os.path.join(config.base_dir,'patchsReshape\\train\\type'+str(i))
    all_files = os.listdir(path_to_write)
    all_files_NDWI = [fn for fn in all_files if fn.startswith(config.GSW_fn) and fn.endswith(config.image_type)]#ndwi.png
    print('ndwi image number:',len(all_files_NDWI))
    for j, fn in enumerate(all_files_NDWI):
        NDWI_img = rasterio.open(os.path.join(path_to_write, fn))
        read_NDWI_img = NDWI_img.read()
        comb_img = np.transpose(read_NDWI_img, axes=(1,2,0)) #Channel at the end  ( , ,1) 
        annotation_im = Image.open(os.path.join(path_to_write, fn.replace(config.GSW_fn,config.annotation_fn)))
        annotation = np.array(annotation_im)
        f = FrameInfo(comb_img, annotation)
        frames.append(f)
    train_img,train_ann= sequential_patches(frames, config.input_image_channel, config.input_label_channel, config.patch_size,config.step_size,augmenter = 'iaa')
    
    train_img_list.append(train_img)
    train_ann_list.append(train_ann)
    print('train patchs number:',len(train_img_list[-1]))
```

```python
def narrayAppend(array1,array2):
    return np.append(array1,array2,axis=0)

def patchSum(patchList):
    if len(patchList)<=1: 
        return patchList[0]
    else:
        sumList= narrayAppend(patchList[0],patchList[1])
        for i in range(2,len(patchList)):
            sumList=narrayAppend(sumList,patchList[i])
#             print(len(sumList))
        return sumList
```

```python
train_img=patchSum(train_img_list)
train_ann=patchSum(train_ann_list)
print("total train data num:",len(train_img))
```

```python
OPTIMIZER = adaDelta
OPTIMIZER = tf.train.experimental.enable_mixed_precision_graph_rewrite(OPTIMIZER)
OPTIMIZER_NAME = 'seq_AdaDelta'
```

```python
OPTIMIZER = adam
OPTIMIZER = tf.train.experimental.enable_mixed_precision_graph_rewrite(OPTIMIZER)
OPTIMIZER_NAME = 'adam'
```

```python
LOSS = tversky 
LOSS_NAME = 'tversky'

# LOSS=focalTversky
# LOSS_NAME = 'focalTversky'

# LOSS=tf.keras.losses.BinaryCrossentropy()
# LOSS_NAME = 'bce'

# LOSS=dice_loss
# LOSS_NAME = 'dice_loss'

# LOSS=bce_dice_loss
# LOSS_NAME = 'bce_dice_loss_28'

timestr = time.strftime("%Y%m%d-%H%M")
chf = config.input_image_channel + config.input_label_channel
chs = reduce(lambda a,b: a+str(b), chf, '') 

if not os.path.exists(config.model_path):
    os.makedirs(config.model_path)
model_path = os.path.join(config.model_path,'lakes_{}_{}_{}_{}_{}.h5'.format(timestr,OPTIMIZER_NAME,LOSS_NAME,chs,config.input_shape[0]))
print(model_path)
chf = config.input_image_channel + config.input_label_channel
chs = reduce(lambda a,b: a+str(b), chf, '') 
```

```python
# Define the model and compile it  
model = UNet([config.BATCH_SIZE, *config.input_shape],config.input_label_channel)
model.compile(optimizer=OPTIMIZER, loss=LOSS, metrics=[dice_coef, dice_loss,accuracy,  precision, recall, IoU])
```

```python
# Define callbacks      for the early stopping of training, LearningRateScheduler and model checkpointing 
from tensorflow.keras.callbacks import ModelCheckpoint, LearningRateScheduler, EarlyStopping, ReduceLROnPlateau, TensorBoard

checkpoint = ModelCheckpoint(model_path, monitor='val_loss', verbose=1, 
                             save_best_only=True, mode='min', save_weights_only = False)

#reduceonplatea： It can be useful when using adam as optimizer
#Reduce learning rate when a metric has stopped improving (after some patience 个epoch, reduce by a factor of 0.33, new_lr = lr * factor). 
#cooldown: number of epochs to wait before resuming normal operation after lr has been reduced. 

reduceLROnPlat = ReduceLROnPlateau(monitor='val_loss', factor=0.33,
                                   patience=4, verbose=1, mode='min',
                                   min_delta=0.0001, cooldown=4, min_lr=1e-16) 

early = EarlyStopping(monitor="val_loss", mode="min", verbose=2, patience=20)


log_dir = os.path.join('./logs','UNet_{}_{}_{}_{}_{}'.format(timestr,OPTIMIZER_NAME,LOSS_NAME,chs, config.input_shape[0]))
tensorboard = TensorBoard(log_dir=log_dir, histogram_freq=0, write_graph=True, write_grads=False, write_images=False, embeddings_freq=0, embeddings_layer_names=None, embeddings_metadata=None, embeddings_data=None, update_freq='epoch')

callbacks_list = [checkpoint, tensorboard, early] #reduceLROnPlat is not required with adaDelta
```

```python
#sequential stratergy
train_img=train_patchs[0]
train_ann=train_patchs[1]
val_img=val_patchs[0]
loss_history = model.fit(train_img, train_ann, 
                         steps_per_epoch=len(train_img)//config.BATCH_SIZE,
                         epochs=config.NB_EPOCHS, 
#                          validation_data=(val_img,val_ann),
                         validation_data=val_patchs,
                         validation_steps=len(val_img)/config.BATCH_SIZE,
                         callbacks=callbacks_list,
                         workers=1,
                         shuffle=True
#                          use_multiprocessing=True # the generator is not very thread safe 
                         #max_queue_size = 60,
                        )
h=loss_history.history
with open('history_{}_{}_{}.txt'.format(OPTIMIZER_NAME,LOSS_NAME,config.input_shape[0]), 'wb') as file_pi:
    pickle.dump(h, file_pi)
plot(h, OPTIMIZER_NAME,LOSS_NAME, config.NB_EPOCHS, config.BATCH_SIZE, config.patch_size[0])
```
